```python
prep_data = data_df.values
prep_data = prep_data[:40000, :]
prep_data.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(prep_data, prep_data)
```

```python
np.save("cos_sim.npy", cosine_sim)
```

```python
def get_recommendations(idx, cosine_sim):
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    sim_scores.sort()
    song_index = [i[0] for i in sim_scores]
    song_index = song_index
    return song_index
```

```python
simm = np.load("cos_sim.npy")
```

```python
song_df["search"] = song_df["name"] + ' ' + song_df["artists"]
song_df.to_csv("song_df.csv", index=False)
```

```python
song_df.index_col = "index"
```

```python
song_df
```

```python
def UseMe(song_input):
    song_index = song_df[song_df.eq(song_input).any(axis=1)].index
    song_index = song_index[0]
    rec_c = get_recommendations(song_index, simm)
    for i in rec_c:
        recomm = song_df.loc[song_df["id"] == i]
        recom = recomm[['name','artists']]
        print(recom)
```

```python
song_input = input("Enter your song")
```
